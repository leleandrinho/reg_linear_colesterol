```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
import scipy.stats as stats
```

```python
dados = pd.read_csv('/content/Colesterol - Colesterol.csv')
```

```python
dados
```

```python
dados_filtrados = dados.query('Total.notnull() and HDL.notnull()')
dados_filtrados
```

```python
x = dados_filtrados[['Total', 'HDL']]
y = dados_filtrados['Idade'].to_numpy()
model = LinearRegression()
model.fit(x, y)
score = model.score(x, y)
print(score)
```

```python
# reconstruindo a coluna HDL
dados_filtrados = dados[ (dados['Idade'].notnull()) & (dados['HDL'].notnull()) & (dados['Total'].notnull()) ]
x = dados_filtrados[['Total']]
y = dados_filtrados['HDL'].to_numpy()
model_HDL = LinearRegression()
model_HDL.fit(x, y)
score = model_HDL.score(x, y)
print(score)
```

```python
def completar_HDL(linha):
  if pd.isna(linha['HDL']) and not pd.isna(linha['Total']):
    # Separa as variáveis regressoras utilizadas no modelo
    campos = linha[['Total']]
    # Retorna para o "apply" o valor previsto
    return round(model_HDL.predict([campos])[0])
  # Se já tem o ano de nascimento, devolve o valor que tem
  return linha['HDL']

# apply: executar a função chamada 1x para cada linha
dados['HDL_calculado'] = dados.apply(completar_HDL, axis=1)
dados
```

```python
# reconstruindo coluna Total
dados_filtrados = dados[ (dados['Idade'].notnull()) & (dados['HDL'].notnull()) & (dados['Total'].notnull()) ]
x = dados_filtrados[['HDL_calculado']]
y = dados_filtrados['Total'].to_numpy()
model_Total = LinearRegression()
model_Total.fit(x, y)
score = model_Total.score(x, y)
print(score)
```

```python
def completar_Total(linha):
  if pd.isna(linha['Total']) and not pd.isna(linha['HDL_calculado']):
    # Separa as variáveis regressoras utilizadas no modelo
    campos = linha[['HDL_calculado']]
    # Retorna para o "apply" o valor previsto
    return round(model_Total.predict([campos])[0])
  # Se já tem o ano de nascimento, devolve o valor que tem
  return linha['Total']

# apply: executar a função chamada 1x para cada linha
dados['Total_calculado'] = dados.apply(completar_Total, axis=1)
dados
```

```python
# calculando regressão com novas colunas
x = dados[['Total_calculado', 'HDL_calculado']]
y = dados['Idade']
model = LinearRegression()
model.fit(x, y)
score = model.score(x, y)
print(score)
```
